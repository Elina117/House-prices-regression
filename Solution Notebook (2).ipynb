```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1 
    df = df[~((df[col] < (Q1 - 2 * IQR)) | (df[col] > (Q3 + 2 * IQR)))]
    return df

for col in filtered_columns:
    df = remove_outliers(df, col)
    
df
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class EncodingClass(BaseEstimator, TransformerMixin):
    
    def __init__(self, categorical, numeric, C=0.005):

        self.categorical = categorical
        self.numeric = numeric
        self.C = C
        
        self.mapping = {}
        
    
    def fit(self, X, y):

        X_fit = X.copy()
        y_fit = y.copy()

        for col in self.categorical:
            mean_value = y_fit.groupby(X_fit[col]).mean()

            noise = self.C * np.random.normal(0, 1, size=len(mean_value))
            self.mapping[col] = mean_value + noise

        return self
        
    def transform(self, df):
        
        df_transformed = df.copy()
        temp = pd.DataFrame()

        for col in self.numeric:
            temp[col] = df_transformed[col]

        for col in self.categorical:
            unique_vals = df_transformed[col].nunique()
            
            if unique_vals > 2:
                if col in self.mapping:
                    temp[col] = df_transformed[col].map(self.mapping[col]).fillna(0)
                else:
                    temp[col] = 0
            else:
                temp = pd.concat([temp, pd.get_dummies(df_transformed[col], prefix=col, drop_first=True)], axis=1).astype('float')

        return temp
```

# Численное решение (с градиентом)

1. Функция потерь (MSE)

   Модель линейной регрессии использует функцию потерь:

   \$
   L(w) = \frac{1}{n} \sum_{i=1}^{n} (X_i w - y_i)^2 = \frac{1}{n} \| Xw - y \|^2
   \$

2. Градиент функции потерь по весам

   Мы хотим вычислить градиент функции потерь по весам \( w \):

   \$
   \frac{\partial L}{\partial w} = \frac{1}{n} \frac{\partial}{\partial w} \left( (Xw - y)^T (Xw - y) \right)
   \$

   \$
   (Xw - y)^T (Xw - y) = w^T X^T X w - 2y^T Xw + y^T y
   \$

3. Производная по весам

   Теперь вычислим производную по весам \( w \):

   \$
   \frac{\partial L}{\partial w} = \frac{1}{n} \left( 2X^T X w - 2X^T y \right)
   \$

```python
import numpy as np

class LinearRegressionGD:
    def __init__(self, learning_rate=0.01, n_iterations=1000, treashold=1e-6):
        
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.treashold = treashold
        self.weights = None
        
        self.history = [] 

    def add_intercept(self, X):
        
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate([intercept, X], axis=1)

    def fit(self, X, y):

        X = self.add_intercept(X)
        self.weights = np.zeros(X.shape[1])

        for i in range(self.n_iterations):
            predictions = X.dot(self.weights)
            errors = predictions - y
            
            gradient = (2 / X.shape[0]) * X.T.dot(errors) # производная от mse по весам
            self.weights -= self.learning_rate * gradient

            mse = np.mean(errors ** 2)
            self.history.append(mse)

            if i > 0 and abs(self.history[i - 1] - self.history[i]) < self.treashold:
                print(f"Остановка на итерации {i} из-за маленького изменения MSE.")
                break

    def predict(self, X):
        
        X = self.add_intercept(X)
        return X.dot(self.weights)

    def get_weights(self):
   
        return self.weights

    def get_history(self):

        return self.history
```

```python
model = LinearRegressionGD(learning_rate=0.01, n_iterations=1000)
model.fit(X_train_scaled, y_train)

predictions_train = model.predict(X_train_scaled)
mse = np.mean((predictions_train - y_train) ** 2)
print(f"Mean Squared Error на трейне: {mse}")

predictions_test = model.predict(X_test_scaled)
mse = np.mean((predictions_test - y_test) ** 2)
print(f"Mean Squared Error на тесте: {mse}")

weights = model.get_weights()
print(f"Коэффициенты модели: {weights}")
```

# Аналитическое решение

\$
w = (X^T X)^{-1} X^T y
\$

К матрице признаков добавляется столбец единиц


```python
import numpy as np

class LinearRegressionAnalytical:
    def __init__(self):
        
        self.coef_ = None
        self.intercept_ = None
    
    def fit(self, X, y):
        
        X_b = np.hstack([np.ones((X.shape[0], 1)), X])
        X_b_T = np.transpose(X_b)
        
        theta = np.dot(np.dot(np.linalg.inv(np.dot(X_b_T, X_b)), X_b_T), y)
        
        self.intercept_ = theta[0]
        self.coef_ = theta[1:]
    
    def predict(self, X):
        X_b = np.hstack([np.ones((X.shape[0], 1)), X])
        return np.dot(X_b, np.concatenate(([self.intercept_], self.coef_)))

```

```python
model = LinearRegressionAnalytical()
model.fit(X_train_scaled, y_train)

predictions_train = model.predict(X_train_scaled)
mse = np.mean((predictions_train - y_train) ** 2)
print(f"MSE на трейне: {mse}")

predictions_test = model.predict(X_test_scaled)
mse = np.mean((predictions_test - y_test) ** 2)
print(f"MSE на тесте : {mse}")

print(f"Коэффициенты: {model.coef_}")
print(f"Свободный член (intercept): {model.intercept_}")
```
